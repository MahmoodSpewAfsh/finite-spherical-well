```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jn, yn

# Define the order of the Bessel function
n = 1

# Define the range for x
x = np.linspace(0, 20, 500)

# Calculate the Bessel functions
y_jn = jn(n, x)
y_yn = yn(n, x)

# Plot the Bessel functions
plt.figure(figsize=(12, 6))

# Plot J_n(x)
plt.subplot(1, 2, 1)
plt.plot(x, y_jn, label=f'J_{n}(x)')
plt.title(f'Bessel Function of the First Kind J_{n}(x)')
plt.xlabel('x')
plt.ylabel(f'J_{n}(x)')
plt.legend()

# Plot Y_n(x)
plt.subplot(1, 2, 2)
plt.plot(x, y_yn, label=f'Y_{n}(x)')
plt.title(f'Bessel Function of the Second Kind Y_{n}(x)')
plt.xlabel('x')
plt.ylabel(f'Y_{n}(x)')
plt.legend()

plt.tight_layout()
plt.show()
```
